```python
# 0. Matplotlib example workflow
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x=[1,2,3,4]
y=[11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(5,5))

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title='Simple plot',
      xlabel='x-axis',
      ylabel='y-axis')

# 5. Save & show (you save the whole figure)
fig.savefig('images/sample-plot.png')
```
